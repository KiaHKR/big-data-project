```python
%matplotlib inline

from pyspark.sql import SparkSession
from pyspark import SparkConf, StorageLevel
import pyspark.sql.functions as F
from cleantext import clean
from pyspark.find_spark_home import _find_spark_home
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from pretty_html_table import build_table
import os

nltk.download('stopwords')
print(_find_spark_home())
```

```python
conf = SparkConf().setMaster('local[*]') \
        .set("spark.local.dir", "F:\spark-temp")
spark = SparkSession.builder.appName("UkraineStuff").config(conf=conf).getOrCreate()
sc = spark.sparkContext
```

```python
df = spark.read.options(header=True, 
                        escape='"', 
                        multiline=True,
                        ).csv('hdfs://localhost:9000/ukraine/Big-Data-Project/extracted/')

min_df = df.drop('_c0', 'userid', 'username', 'followers', 'acctdesc', 'location', 'following', 'totaltweets', 'usercreatedts', 'tweetid', 'tweetcreatedts', 'retweetcount', 'coordinates', 'favorite_count', 'extractedts')

min_df.repartition(256).persist(StorageLevel.MEMORY_AND_DISK).write.mode('overwrite').parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/results')
```

```python
min_df = spark.read.parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/results')

df_index = min_df.select("*").withColumn("id", F.monotonically_increasing_id())
column_list = df_index.columns
column_list.remove('id')

df_index = df_index.select('id', *column_list)

rows_to_delete = df_index.filter(df_index['language'] != 'en')

df_index = df_index.join(rows_to_delete, on=['id'], how='left_anti')

df_index.repartition(256, 'id').persist(StorageLevel.MEMORY_AND_DISK).write.mode('overwrite').parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/english_removed')
```

```python
df_index = spark.read.parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/english_removed')

@F.udf
def clean_emojis_and_urls(s):
    if s is not None:
        return clean(s, no_emoji=True, no_urls=True, no_line_breaks=True, no_numbers=True, replace_with_number='', no_punct=True, no_currency_symbols=True, replace_with_url="", replace_with_currency_symbol="")

df_index = df_index.withColumn('text', F.regexp_replace('text', r'[#][a-zA-Z0-9_#]*', ''))
df_index = df_index.withColumn('text', F.regexp_replace('text', r'[@]', ''))
df_index = df_index.withColumn('text', F.regexp_replace('text', r"['’‘]s", ''))
df_index = df_index.withColumn('text', clean_emojis_and_urls('text'))
df_index = df_index.withColumn('text', F.regexp_replace('text', r'<url>', ''))

df_index.repartition(256, 'id').persist(StorageLevel.MEMORY_AND_DISK).write.mode('overwrite').parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/text_cleaned')
```

```python
df_index = spark.read.parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/text_cleaned')

rows_to_delete = df_index.filter(df_index['text'] == '')

df_index = df_index.join(rows_to_delete, on=['id'], how='left_anti')

df_index.repartition(256, 'id').persist(StorageLevel.MEMORY_AND_DISK).write.mode('overwrite').parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/empty_text_removed')
```

```python
df_index = spark.read.parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/empty_text_removed')

clean_df = df_index.withColumn("words", F.explode(F.split(F.trim("text"), " "))).groupBy("words").count()

# Remove stopwords
stop_words = set(stopwords.words('english'))
rows_to_delete = clean_df.filter(clean_df.words.isin(stop_words))
clean_df = clean_df.join(rows_to_delete, on=['words'], how='left_anti')

rows_to_delete = clean_df.filter(clean_df['count'] <= 50)
clean_df = clean_df.join(rows_to_delete, on=['words'], how='left_anti')

clean_df.persist(StorageLevel.MEMORY_AND_DISK).write.mode('overwrite').parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/final_count')
```

```python
clean_df = spark.read.parquet('hdfs://localhost:9000/ukraine/Big-Data-Project/final_count')

clean_df = clean_df.orderBy(F.desc('count'))

words = clean_df.rdd.map(lambda x: (x[0], x[1])).collectAsMap()
```

```python
html_table = clean_df.toPandas().to_html()

try:
    os.remove(os.path.join(os.path.abspath(''), 'index.html'))
except OSError:
    pass


with open(os.path.join(os.path.abspath(''), 'index.html'), 'x') as f:
    f.write(html_table)
```
